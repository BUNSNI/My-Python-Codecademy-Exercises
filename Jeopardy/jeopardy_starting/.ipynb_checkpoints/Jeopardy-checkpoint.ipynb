## Printing out the column names, we can see that, apart from the first, each one has a space at the front. Let's rename the columns:

```python
for col in jeopardy_df.columns:
    jeopardy_df.rename(columns= {col : col.strip()}, inplace=True)
print(jeopardy_df.columns)
```

## Create a new column with values without the '$' character

```python
jeopardy_df['Clean_Values'] = jeopardy_df['Value'].map(lambda x: x.strip('$,'))
```

## Convert new values in new column to floats

## We may want to eventually compute aggregate statistics, like .mean() on the " Value" column. But right now, the values in that column are strings. Convert the " Value" column to floats. If you’d like to, you can create a new column with the float values.

```python
jeopardy_df['Clean_Values'] = pd.to_numeric(jeopardy_df['Clean_Values'], errors='coerce')
```

## Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list ["King", "England"] was passed to our function, the function returned a DataFrame of 152 rows.

### This is my first attempt:

```python
def find_question(words):
    num = 0
    total_value = 0
    for row in jeopardy_df.itertuples():
        if all(x.lower() in row.Question.lower() for x in words):
            print(row.Question)
            num += 1
            if pd.isna(row.Clean_Values):
                pass
            else:
                total_value += row.Clean_Values
    average = total_value / num
    print('We have found ' + str(num) + ' questions that contain the words you are looking for.')
    print('The mean value of questions that contain the word or words: ' + words[0] + ', ' + words[1] + ' is: $ ' + str(int(average)))
            

find_question(['King', 'England'])
```

## Now that you can filter the dataset of question, use your new column that contains the float values of each question to find the “difficulty” of certain topics. For example, what is the average value of questions that contain the word "King"?

```python
def search_question(arg1):
    if len(arg1) == 1:
        first_word = jeopardy_df[jeopardy_df['Question'].str.contains(arg1[0], case=False)]
    
        print('We have found ' + str(len(first_word)) + ' questions that contain the words you are looking for.')
        print('The mean value of questions that contain the words: ' + arg1[0] + ' is: $' + str(first_word.Clean_Values.mean()))
        print(first_word.Question)   
        
        
    else:
        first_word = jeopardy_df[jeopardy_df['Question'].str.contains(arg1[0], case=False) & jeopardy_df['Question'].str.contains(arg1[len(arg1)-1], case=False)]
   
        print('We have found ' + str(len(first_word)) + ' questions that contain the words you are looking for.')
        print('The mean value of questions that contain the words: ' + arg1[0] + ' and ' + arg1[1] + ' is: $' + str(first_word.Clean_Values.mean()))
        print(first_word.Question)

search_question(['king', 'England'])
```

## Write a function that returns the count of the unique answers to all of the questions in a dataset. For example, after filtering the entire dataset to only questions containing the word "King", we could then find all of the unique answers to those questions. The answer “Henry VIII” appeared 3 times and was the most common answer.

```python
def search_question(arg1):
    if len(arg1) == 1:
        first_word = jeopardy_df[jeopardy_df['Question'].str.contains(arg1[0], case=False)]
    
        print('We have found ' + str(len(first_word)) + ' questions that contain the words you are looking for.')
        print('The mean value of questions that contain the words: ' + arg1[0] + ' is: $' + str(first_word.Clean_Values.mean()))
        print(first_word.Question)
        print(first_word.Answer.value_counts())
        
        
    else:
        first_word = jeopardy_df[jeopardy_df['Question'].str.contains(arg1[0], case=False) & jeopardy_df['Question'].str.contains(arg1[len(arg1)-1], case=False)]
   
        print('We have found ' + str(len(first_word)) + ' questions that contain the words you are looking for.')
        print('The mean value of questions that contain the words: ' + arg1[0] + ' and ' + arg1[1] + ' is: $' + str(first_word.Clean_Values.mean()))
        print(first_word.Question)
        print(first_word.Answer.value_counts())

search_question(['king', 'England'])
```

## Investigate the ways in which questions change over time by filtering by the date. How many questions from the 90s use the word "Computer" compared to questions from the 2000s?

```python
computer_90s = jeopardy_df[(jeopardy_df['Air Date'] > '1989-12-31') & (jeopardy_df['Air Date'] < '2000') & jeopardy_df['Question'].str.contains('Computer')]
computer_after_90s = jeopardy_df[(jeopardy_df['Air Date'] > '1999-12-31') & jeopardy_df['Question'].str.contains('Computer')]
print(len(computer_90s))
print(computer_90s)
print(len(computer_after_90s))
print(computer_after_90s)
```
